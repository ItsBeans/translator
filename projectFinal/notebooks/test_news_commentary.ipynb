```python
from datasets import load_dataset

dataset = load_dataset("Helsinki-NLP/news_commentary", "en-es", split = 'train')
```

```python
checkpoint = "beanslmao/hensinki-es-en-finetuned-spanish-to-english-tateoba"

import evaluate

metric = evaluate.load("sacrebleu")

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(checkpoint)

from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer

model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)

from transformers import pipeline

translator = pipeline("translation", model=model, tokenizer=tokenizer)
```

```python
cleanData = dataset.select_columns('translation')

print(cleanData[450]['translation']['en'])

```

```python
from nltk.translate.bleu_score import sentence_bleu
import string


engToSpanish = cleanData[450]['translation']['es']

engToSpanish = translator(engToSpanish)
engToSpanish = (engToSpanish[0]).get('translation_text')
engToSpanish = engToSpanish.lower()
print(engToSpanish)
translatorTable = str.maketrans('', '', string.punctuation)
engToSpanish = engToSpanish.translate(translatorTable)
print(engToSpanish)
```

```python
from nltk.translate.bleu_score import sentence_bleu
import string

predictions = []

for x in range(1000):
  temp = cleanData[x]['translation']['es']
  temp = translator(temp)
  temp = (temp[0]).get('translation_text')
  temp = temp.lower()
  translatorTable = str.maketrans('', '', string.punctuation)
  temp = temp.translate(translatorTable)

  predictions.append(temp)
  print('number:',x)

print(predictions)
```

```python
references = []

for x in range(1000):
  temp = []
  temp = [cleanData[x]['translation']['en']]
  references.append(temp)


print(references)
results = metric.compute(predictions=predictions,references=references)
print(results)
```
