```python
from datasets import load_dataset

dataset_english = load_dataset("AmazonScience/massive", "en-US", split='test')
dataset_spanish = load_dataset("AmazonScience/massive", "es-ES", split='test')
```

```python
print(dataset_english[2]['utt'])
```

```python
checkpoint = "Helsinki-NLP/opus-mt-en-es"

import evaluate

metric = evaluate.load("sacrebleu")

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(checkpoint)

from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer

model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)

from transformers import pipeline

translator = pipeline("translation", model=model, tokenizer=tokenizer)
```

```python
from sacrebleu.metrics import BLEU
bleu = BLEU()
```

```python
newEnglish = dataset_english.select_columns('utt')
newSpanish = dataset_spanish.select_columns('utt')
print(newEnglish['utt'])
print(newSpanish['utt'])
print(len(newEnglish))
```

```python
from nltk.translate.bleu_score import sentence_bleu
import string


temp = newSpanish[450]['utt']
temp = translator(temp)
temp = (temp[0]).get('translation_text')
temp = temp.lower()
print(temp)
translatorTable = str.maketrans('', '', string.punctuation)
temp = temp.translate(translatorTable)
print(temp)
```

```python
from nltk.translate.bleu_score import sentence_bleu
import string

predictions = []

for x in range(len(newEnglish)):
  temp = newEnglish[x]['utt']
  temp = translator(temp)
  temp = (temp[0]).get('translation_text')
  temp = temp.lower()
  translatorTable = str.maketrans('', '', string.punctuation)
  temp = temp.translate(translatorTable)

  predictions.append(temp)
  print('number:',x)

print(predictions)
```

```python
references = []

for x in range(len(newSpanish)):
  temp = []
  temp = [newSpanish[x]['utt']]
  references.append(temp)


print(references)
results = metric.compute(predictions=predictions,references=references)
print(results)
```
