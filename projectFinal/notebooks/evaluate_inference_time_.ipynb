```python
checkpoint = "Helsinki-NLP/opus-mt-en-es"

from transformers import AutoTokenizer
from transformers import pipeline
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
import time

model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
translator = pipeline("translation", model=model, tokenizer=tokenizer)
```

```python
word10 = 'Hello my name is Benny. This sentence is very short'
word5 = 'Hello my name is Benny'
word20 = 'Hello my name is Benny This sentence is very short Hello my name is Benny This sentence is very short'
word50 = 'great thin weapon react lung emergency shoe none subject here grocery ready orientation wear white hunter display register video additional greatest mission advantage refugee constitutional message degree rush wine necessary put see wing matter size widespread living tourist practice wheel politician captain load vegetable English latter deficit occur him and'
```

```python
score = 0

for x in range (10):
  timeStart = time.time()
  translator(word50)
  timeEnd = time.time()
  score = score + (timeEnd - timeStart)


print(score)
print(score/10)
```
